```python
import os
import warnings
import yaml
from pathlib import Path

PROJECT_DIR = Path().resolve()
PROJECT_ROOT_NAME = 'spectroscopy-regression'

while PROJECT_DIR.name != PROJECT_ROOT_NAME and PROJECT_DIR != PROJECT_DIR.parent:
    PROJECT_DIR = PROJECT_DIR.parent

if Path().resolve() != PROJECT_DIR:
    os.chdir(PROJECT_DIR)
```

```python
import pandas as pd
import os

input_dir = "data/NGC_combi.csv"      
columns_to_keep = ['Na/Fe', 'F275W_abs', 'F336W_abs', 'F435W_abs', 'F606W_abs', 'F814W_abs', 
                   'F275W', 'F336W', 'F435W', 'F606W', 'F814W', 'age_Kruijssen', 'Fe/H']

df = pd.read_csv(input_dir, usecols=columns_to_keep)
for col in columns_to_keep:
    df[col] = pd.to_numeric(df[col], errors='coerce')

mask = (
    (df < -100).any(axis=1) |         
    df.isna().any(axis=1)             
)
df = df[~mask] 
df.to_csv('data/Photometry+Age+Metallicity+NaFe.csv', index=False)
```
